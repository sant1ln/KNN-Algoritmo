```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

data = {'Masa':   [65,      68,      70,         65,      69,     73,     75,      68,      70,         60,      54,     55, 70,65,      68,      70,         45,      80,     43,     55,      48,      89,         61,      80,     55, 70],
        'Altura': [1.70,    1.45,    1.75,       1.70,    1.60,   1.63,   1.70,    1.65,    1.80,       1.60,    1.63,   1.62,1.70,1.70,    1.45,    1.75,       1.66,    1.60,   1.63,   1.50,    1.41,    1.50,       1.65,    1.43,   1.62,1.70],
        'Genero': ['Hombre','Mujer', 'Hombre',  'Mujer',  'Mujer','Hombre','Hombre','Mujer', 'Hombre',  'Mujer',  'Mujer','Hombre','Mujer', 'Hombre','Mujer', 'Hombre',  'Hombre',  'Mujer','Hombre','Hombre','Mujer', 'Hombre',  'Mujer',  'Mujer','Hombre','Mujer']}
punto_nuevo = {'Masa': [80],
               'Altura': [1.65]}
df = pd.DataFrame(data)
punto_nuevo = pd.DataFrame(punto_nuevo)
ax = plt.axes()
ax.scatter(df.loc[df['Genero'] == 'Hombre', 'Masa'],
           df.loc[df['Genero'] == 'Hombre', 'Altura'],
           c="red",
           label="Hombre")
ax.scatter(df.loc[df['Genero'] == 'Mujer', 'Masa'],
           df.loc[df['Genero'] == 'Mujer', 'Altura'],
           c="blue",
           label="Mujer")
ax.scatter(punto_nuevo['Masa'],
           punto_nuevo['Altura'],
           c="yellow")
plt.xlabel("Masa")
plt.ylabel("Altura")
ax.legend()
plt.show()

knn = KNeighborsClassifier(n_neighbors=11)
X = df[['Masa', 'Altura']]
y = df[['Genero']]
knn.fit(X, y)
prediccion = knn.predict(punto_nuevo)
print(prediccion)
```
